```python
#dividing dependent and independent features
```

```python
X = df.iloc[:,:-1] #independent features
```

```python
Y = df.iloc[:,-1]
```

```python
#Dividing of train and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, train_size = 0.38, random_state = 45)
```

```python
#feature scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.fit(X_test)
```

```python
#Training the model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regression = LogisticRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
#predcit the values
```

```python
y_pred = regression.predict(X_test)
```

```python
#Performance metrics
```

```python
#confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
con_mat = confusion_matrix(Y_test,y_pred)
```

```python
print(con_mat)
```

```python
true_positive = con_mat[0][0]
false_positive = con_mat[0][1]
false_negative = con_mat[1][0]
true_negative = con_mat[1][1]
```

```python
Accuracy = (true_positive+true_negative)/(true_positive+false_positive+false_negative+true_negative)
```

```python
print(Accuracy)
```

```python
Precision = true_positive/(true_positive+false_positive)
Precision
```

```python
Recall = true_positive/(true_positive+false_negative)
Recall
```

```python
F1_Score = 2*(Recall * Precision) / (Recall + Precision)
F1_Score
```
